```python
import cv2
import numpy as np

# Load an image
image = cv2.imread('street.jpg')

# Convert to HSV color space
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Define the range for blue color
lower_blue = np.array([100, 150, 0])   # Lower bound (H, S, V)
upper_blue = np.array([140, 255, 255]) # Upper bound (H, S, V)

# Create a binary mask
mask = cv2.inRange(hsv_image, lower_blue, upper_blue)
from google.colab.patches import cv2_imshow
# Display the mask
cv2_imshow( mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


# Load YOLO
net = cv2.dnn.readNet("yolov4.weights", "yolov4.cfg")
layer_names = net.getLayerNames()
output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

# Load COCO class labels (used by YOLOv4)
with open("coco.names", "r") as f:
    classes = [line.strip() for line in f.readlines()]

# Load image
image = cv2.imread("drive/MyDrive/Colab Notebooks/street.jpg")
height, width, channels = image.shape

# Preprocess the image for YOLO
blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
net.setInput(blob)
outs = net.forward(output_layers)

# Initialize lists for detected objects
class_ids = []
confidences = []
boxes = []

# Process YOLO outputs
for out in outs:
    for detection in out:
        scores = detection[5:]
        class_id = np.argmax(scores)
        confidence = scores[class_id]
        if confidence > 0.5:  # Confidence threshold
            center_x = int(detection[0] * width)
            center_y = int(detection[1] * height)
            w = int(detection[2] * width)
            h = int(detection[3] * height)
            x = int(center_x - w / 2)
            y = int(center_y - h / 2)
            boxes.append([x, y, w, h])
            confidences.append(float(confidence))
            class_ids.append(class_id)

# Apply Non-Maximum Suppression to eliminate redundant boxes
indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

# Draw bounding boxes and add labels with confidence
for i in range(len(boxes)):
    if i in indexes:
        x, y, w, h = boxes[i]
        label = str(classes[class_ids[i]])  # Get class label
        confidence = confidences[i]  # Get confidence score
        color = (255, 0, 0)  # Set a color for the box
        cv2.rectangle(image, (x, y), (x + w, y + h), color, 1)
        cv2.putText(image, f"{label} {confidence:.2f}", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.4, color, 1)

# Display the image with bounding boxes and labels
cv2_imshow( image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
